# 1. Álgebra de Matrices en Analítica de Datos

## **Importancia del álgebra de matrices en ciencia de datos**

El álgebra de matrices permite representar de forma compacta y operativa grandes volúmenes de información estructurada. En analítica de datos, un conjunto de observaciones con múltiples variables puede representarse como una **matriz de datos**, donde:

- Cada **fila** representa una observación o instancia.
- Cada **columna** representa una variable o característica.

A partir de esta representación, el álgebra de matrices facilita operaciones fundamentales como la **transformación de datos**, el **cálculo de métricas multivariadas**, la **reducción de dimensionalidad** o la **formulación y resolución de modelos lineales**, entre muchas otras.

---

## **Operaciones básicas con matrices**

Al trabajar con matrices de datos se aplican operaciones como:

- **Transposición (`Aᵗ`)**: útil para reorganizar datos o expresar correlaciones.
- **Multiplicación matricial**: base de los modelos lineales (`X · β = y`).
- **Inversa o pseudoinversa**: se emplea para resolver sistemas o ajustar modelos.
- **Suma y resta de matrices**: para combinar o comparar conjuntos de datos.

---

## **Aplicación práctica: predicción de ingresos a partir de variables socioeconómicas**

En este ejemplo simulamos un pequeño conjunto de datos con tres variables: *años de educación*, *horas trabajadas por semana*, y *experiencia laboral (en años)*, con el objetivo de predecir los *ingresos estimados* de una persona. Modelaremos este caso como una regresión lineal matricial.


```python
import numpy as np

# --- Matriz de características (X) ---
# Columnas: [Educación, Horas/semana, Experiencia]
X = np.array([
    [12, 40, 5],
    [16, 45, 7],
    [10, 38, 3],
    [14, 50, 10],
    [13, 42, 6]
])

# --- Vector objetivo (y): Ingresos observados en miles ---
y = np.array([[35], [50], [30], [60], [40]])

# --- Agregamos término de sesgo (bias) a X ---
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # añade columna de 1s

# --- Cálculo de coeficientes óptimos usando la fórmula matricial normal ---
# β = (XᵗX)^(-1)Xᵗy
X_transpose = X_b.T
beta = np.linalg.inv(X_transpose @ X_b) @ X_transpose @ y

print("Coeficientes del modelo (intercepto + pesos):")
print(beta)

# --- Predicciones del modelo ---
y_pred = X_b @ beta
print("\nPredicciones de ingresos (en miles):")
print(y_pred)

# --- Comparación con valores reales ---
print("\nValores reales:")
print(y)
```

Este modelo realiza una predicción lineal multivariable de ingresos a partir de características socioeconómicas mediante operaciones matriciales fundamentales. Así se demuestra cómo el álgebra de matrices no solo representa datos, sino que también constituye el núcleo computacional de los modelos estadísticos en ciencia de datos.

<p align="justify">
Utilice un conjunto de datos para calcular valores y vectores propios de una matriz asociada.
Explique cómo estos resultados pueden influir en el diseño de indicadores de impacto.</p>

<p align="justify">
El análisis de valores y vectores propios aplicado a matrices derivadas de conjuntos de datos representa una herramienta fundamental para la reducción dimensional y la identificación de patrones subyacentes en estudios de impacto social o económico. A partir de una matriz de covarianza o correlación construida con variables representativas de un fenómeno (educación, pobreza, empleo, etc.), es posible calcular los valores propios (eigenvalues) y vectores propios (eigenvectors). Estos elementos permiten determinar qué combinaciones lineales de variables explican la mayor parte de la variabilidad del sistema observado.</p>
<p align="justify">
Los valores propios cuantifican la proporción de varianza explicada por cada componente, mientras que los vectores propios definen las direcciones principales del espacio de datos, ayudando a identificar conglomerados de variables que responden de manera conjunta ante intervenciones o cambios. Esta técnica resulta especialmente útil al diseñar indicadores compuestos, ya que permite seleccionar variables relevantes, eliminar redundancias y construir indicadores más sintéticos, explicativos y robustos frente a variaciones contextuales. En conjunto, este análisis contribuye a la creación de herramientas más precisas para el monitoreo y la evaluación del impacto en políticas públicas o programas de intervención.</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Simulación de un conjunto de datos
np.random.seed(42)
data = {
    "AsistenciaEscolar": np.random.normal(80, 10, 100),
    "EducacionPadres": np.random.normal(12, 2, 100),
    "RecursosEscolares": np.random.normal(50, 8, 100),
    "AccesoInternet": np.random.normal(70, 12, 100)
}

df = pd.DataFrame(data)

# Escalado de datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Cálculo de componentes principales
pca = PCA()
pca.fit(df_scaled)

# Valores propios
eigenvalues = pca.explained_variance_
print("Valores propios (varianza explicada por componente):")
print(eigenvalues)

# Vectores propios
eigenvectors = pca.components_
print("\nVectores propios (componentes principales):")
print(eigenvectors)

# Gráfico de varianza explicada
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title("Varianza explicada acumulada")
plt.xlabel("Número de componentes")
plt.ylabel("Proporción de varianza explicada")
plt.grid()
plt.show()
```

# **Aplicación del producto escalar y vectorial en el filtrado y limpieza de datos**
El **producto escalar** y el producto vectorial son operaciones fundamentales del álgebra lineal, con aplicaciones directas en el procesamiento de datos. En el contexto de análisis de datos bidimensionales, estas operaciones permiten identificar relaciones entre vectores de características y aplicar técnicas de proyección para filtrar o depurar datos ruidosos o irrelevantes.
El producto escalar (también conocido como producto punto) permite cuantificar el grado de alineación entre dos vectores. Esto resulta útil, por ejemplo, al proyectar datos sobre una dirección dada para identificar su componente útil (relevante) y separar el componente ortogonal (ruido). Así, se pueden generar filtros basados en umbrales angulares o energéticos.

Por su parte, el **producto vectorial** en espacios bidimensionales se interpreta como el área del paralelogramo formado por dos vectores en el plano. En el análisis de datos, esta magnitud puede usarse para detectar cambios en la orientación o consistencia direccional entre datos vecinos, útil por ejemplo para detección de anomalías o inconsistencias estructurales en registros.

En conjunto, ambos productos permiten realizar **proyecciones, filtrados angulares y detección de valores atípicos** de manera precisa, lo cual resulta altamente relevante en etapas de preprocesamiento y limpieza de bases de datos.

Ejemplo práctico de producto escalar, vectorial y proyecciones para limpieza de datos 2D:


```python
import numpy as np
import matplotlib.pyplot as plt

# Generación de un conjunto de datos bidimensional con un poco de "ruido"
np.random.seed(0)
n = 100
x_dir = np.array([1, 2])  # Dirección principal (puede representar una variable relevante)
x_dir = x_dir / np.linalg.norm(x_dir)  # Normalizamos

# Datos alineados + ruido perpendicular
datos = np.outer(np.linspace(0, 10, n), x_dir)  # Datos alineados a x_dir
ruido_perp = np.random.normal(0, 0.5, size=(n, 2))
ruido_perp = ruido_perp - (ruido_perp @ x_dir).reshape(-1,1) * x_dir  # Solo componente perpendicular
datos_ruidosos = datos + ruido_perp

# Proyección escalar sobre la dirección relevante
proyecciones_escalar = datos_ruidosos @ x_dir
proyecciones_vector = np.outer(proyecciones_escalar, x_dir)

# Limpieza: se proyectan datos ruidosos sobre la dirección relevante
datos_limpios = proyecciones_vector

# Visualización
plt.figure(figsize=(8,6))
plt.quiver(np.zeros(n), np.zeros(n), datos_ruidosos[:,0], datos_ruidosos[:,1],
           angles='xy', scale_units='xy', scale=1, color='lightgray', label='Datos con ruido')
plt.quiver(np.zeros(n), np.zeros(n), datos_limpios[:,0], datos_limpios[:,1],
           angles='xy', scale_units='xy', scale=1, color='blue', label='Datos proyectados (limpios)')
plt.plot(x_dir[0]*10, x_dir[1]*10, 'ro', label='Dirección base')
plt.title("Proyección escalar para limpieza de datos")
plt.axis("equal")
plt.grid()
plt.legend()
plt.show()
```

El anterior ejemplo simula un conjunto de datos bidimensionales con ruido perpendicular a una dirección significativa, y aplica el producto escalar para realizar la proyección y eliminar el componente no deseado. La diferencia visual entre los datos ruidosos y los proyectados ilustra cómo estas herramientas ayudan en la limpieza y manipulación de datos.

**Optimización numérica y métodos computacionales aplicados en problemas reales de ciencia de datos**

La optimización numérica es un pilar fundamental en el análisis computacional de problemas de ciencia de datos. Muchos algoritmos, desde el ajuste de modelos predictivos hasta la calibración de parámetros en sistemas complejos, se basan en la identificación de mínimos o máximos de funciones objetivo. A través de métodos numéricos como gradiente descendente, descenso estocástico, Newton-Raphson o algoritmos genéticos, es posible encontrar soluciones aproximadas a problemas cuya forma analítica es inaccesible o ineficiente de resolver simbólicamente.

Un caso representativo es la regresión lineal, donde el objetivo es minimizar el error cuadrático medio entre los valores predichos y los observados. Esta minimización se puede abordar mediante métodos computacionales iterativos que ajustan los parámetros del modelo de forma eficiente. Tales técnicas no solo permiten mejorar el rendimiento de los modelos, sino que también permiten adaptarse a conjuntos de datos grandes, dinámicos y ruidosos, características comunes en escenarios reales de producción.

En suma, la aplicación de métodos de optimización computacional no solo incrementa la calidad de los modelos predictivos, sino que también habilita soluciones más rápidas, adaptativas y escalables para problemas reales del análisis de datos.

En el siguiente ejemplo usamos gradiente descendente para minimizar una función de error cuadrático en una regresión lineal con datos simulados:


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados (relación lineal con ruido)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 5 + 3 * X + np.random.randn(100, 1)

# Añadimos un vector de 1s para el término de sesgo (bias)
X_b = np.c_[np.ones((100, 1)), X]

# Inicialización de parámetros (theta)
theta = np.random.randn(2,1)

# Función de costo
def compute_mse(X_b, y, theta):
    predictions = X_b @ theta
    mse = np.mean((predictions - y) ** 2)
    return mse

# Gradiente descendente
eta = 0.1       # tasa de aprendizaje
n_iterations = 100
m = len(X_b)
mse_history = []

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T @ (X_b @ theta - y)
    theta = theta - eta * gradients
    mse_history.append(compute_mse(X_b, y, theta))

# Resultados
print("Parámetros optimizados (theta):")
print(theta)

# Visualización de la función de costo (convergencia)
plt.plot(mse_history)
plt.xlabel("Iteración")
plt.ylabel("Error cuadrático medio (MSE)")
plt.title("Convergencia por gradiente descendente")
plt.grid()
plt.show()

# Visualización de los datos y la recta ajustada
plt.scatter(X, y, color="gray", label="Datos")
plt.plot(X, X_b @ theta, color="red", label="Modelo ajustado")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regresión lineal por optimización numérica")
plt.legend()
plt.show()
```

Este ejemplo muestra cómo una técnica de optimización computacional como gradiente descendente permite encontrar los parámetros óptimos de un modelo lineal a partir de datos con ruido. El enfoque puede escalarse a redes neuronales, sistemas logísticos, modelos de clasificación y más.

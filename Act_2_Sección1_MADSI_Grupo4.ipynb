### H1 · Rendimiento 1er semestre vs Deserción
Crearemos métricas de rendimiento del primer semestre y compararemos su distribución entre estados de `Target` (Dropout, Enrolled, Graduate).  
Métricas:
- `pass_rate_1st` = aprobadas / inscritas (1er semestre)
- `grade_1st` = nota media del 1er semestre (ya existe en la columna)
Luego veremos estadísticas por `Target`, tablas y gráficos.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- 1) Crear métricas del 1er semestre ---
df = df.copy()

# Evitar división por cero
enrolled_1st = df['Curricular units 1st sem (enrolled)'].replace(0, np.nan)
df['pass_rate_1st'] = df['Curricular units 1st sem (approved)'] / enrolled_1st
df['grade_1st'] = df['Curricular units 1st sem (grade)']  # alias corto

# Bandera de deserción (booleano)
df['is_dropout'] = (df['Target'] == 'Dropout').astype(int)

print(df[['pass_rate_1st','grade_1st','Target','is_dropout']].head())
````

### Tasa de deserción por tramos de rendimiento

Usamos **Kruskal–Wallis** para contrastar si las distribuciones de `grade_1st` y `pass_rate_1st` difieren entre grupos de `Target`.

## Hipótesis 2: Relación entre becas, deudas y deserción estudiantil

En esta sección analizaremos si la condición de ser beneficiario de una beca o tener deudas con la institución influye en la probabilidad de deserción.

**Objetivos específicos:**
1. Calcular la tasa de deserción entre estudiantes con y sin beca.
2. Calcular la tasa de deserción entre estudiantes con y sin deudas.
3. Analizar la combinación de ambas variables (beca y deuda) para detectar patrones de riesgo.
4. Visualizar los resultados para identificar tendencias de forma más clara.
5. Realizar una prueba estadística para evaluar si las diferencias observadas son significativas.

Este análisis nos permitirá identificar si estas variables financieras están asociadas a un mayor o menor riesgo de abandono académico.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# --- 1. Tasa de deserción según beca ---
scholarship_dropout = df.groupby('Scholarship holder')['Target'].apply(lambda x: (x == 'Dropout').mean())
print("📌 Tasa de deserción según beca:")
print(scholarship_dropout)

# --- 2. Tasa de deserción según deuda ---
debt_dropout = df.groupby('Debtor')['Target'].apply(lambda x: (x == 'Dropout').mean())
print("\n📌 Tasa de deserción según deuda:")
print(debt_dropout)

# --- 3. Combinación de beca y deuda ---
combo_dropout = df.groupby(['Scholarship holder', 'Debtor'])['Target'].apply(lambda x: (x == 'Dropout').mean())
print("\n📌 Tasa de deserción por combinación beca/deuda:")
print(combo_dropout)

# --- 4. Visualización ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x='Scholarship holder', y=(df['Target'] == 'Dropout').astype(int), data=df, ax=axes[0])
axes[0].set_title('Tasa de deserción según beca')
axes[0].set_ylabel('Proporción de deserción')

sns.barplot(x='Debtor', y=(df['Target'] == 'Dropout').astype(int), data=df, ax=axes[1])
axes[1].set_title('Tasa de deserción según deuda')
axes[1].set_ylabel('Proporción de deserción')

plt.tight_layout()
plt.show()

# --- 5. Prueba estadística (Chi-cuadrado) ---
# Para beca
contingency_scholar = pd.crosstab(df['Scholarship holder'], df['Target'] == 'Dropout')
chi2_sch, p_sch, _, _ = chi2_contingency(contingency_scholar)

# Para deuda
contingency_debt = pd.crosstab(df['Debtor'], df['Target'] == 'Dropout')
chi2_debt, p_debt, _, _ = chi2_contingency(contingency_debt)

print(f"\n📊 Chi-cuadrado beca: χ²={chi2_sch:.3f}, p-value={p_sch:.3e}")
print(f"📊 Chi-cuadrado deuda: χ²={chi2_debt:.3f}, p-value={p_debt:.3e}")
````

**2. Efecto de las deudas:**
- Los estudiantes **con deuda** tienen una tasa de deserción de **62%**, comparado con **28%** en quienes no tienen deuda.

**3. Combinación beca + deuda:**
- El grupo más vulnerable es **sin beca y con deuda**, con una tasa de deserción de **67%**.
- El grupo más protegido es **con beca y sin deuda**, con una tasa de deserción de apenas **10%**.

**Conclusión:**
Las becas actúan como un factor protector, mientras que las deudas se asocian a un incremento marcado en el riesgo de abandono. La combinación de ambos factores (falta de beca + deuda) parece ser el escenario de mayor vulnerabilidad estudiantil.

   - **Kruskal–Wallis**: diferencias significativas entre grupos para notas (`H=1390.744`, `p<0.001`) y tasas de aprobación (`H=2324.516`, `p<0.001`).  
   - **Chi-cuadrado**: asociación significativa entre tramos de nota (`χ²=1798.289`, `p<0.001`) y de tasa de aprobación (`χ²=2677.001`, `p<0.001`) con el estado `Target`.

**Interpretación:**  
El rendimiento académico en el segundo semestre está fuertemente relacionado con la deserción. Los estudiantes con bajas calificaciones y tasas de aprobación reducidas tienen un riesgo muy elevado de abandonar, mientras que quienes presentan un buen rendimiento tienen una probabilidad significativamente menor de deserción.

**Implicación práctica:**  
El seguimiento temprano de las métricas del segundo semestre permitiría implementar acciones preventivas inmediatas, enfocadas especialmente en estudiantes con rendimiento bajo o en declive, para disminuir las tasas de deserción.


## Conclusiones Generales

El análisis realizado tuvo como propósito identificar los factores académicos, socioeconómicos y contextuales que inciden en la deserción estudiantil, con el fin de orientar estrategias institucionales de retención. A partir de la hipótesis inicial, se evaluaron cinco hipótesis específicas.

### 1. Rendimiento académico en el primer semestre (H1)
Se evidenció que un bajo rendimiento inicial —medido por la nota promedio y la tasa de aprobación— está fuertemente asociado con la deserción. Los estudiantes con notas bajas o sin aprobación de materias presentaron tasas de *Dropout* significativamente mayores. Esto sugiere que el seguimiento temprano al desempeño del primer semestre es crítico para activar intervenciones.

### 2. Presencia de deuda (H2)
La condición de *Debtor* mostró una relación directa con la deserción: los estudiantes con deuda activa tuvieron mayor probabilidad de abandonar los estudios. Esto refleja la importancia de integrar medidas de apoyo financiero y planes de pago flexibles como parte de las estrategias de retención.

### 3. Beneficiarios de beca (H3)
Los estudiantes con beca presentaron menor probabilidad de deserción, lo que confirma el papel positivo de las ayudas económicas en la permanencia académica. Sin embargo, este efecto no es absoluto y debe complementarse con acompañamiento académico y psicosocial.

### 4. Modalidad de asistencia (H4)
La modalidad de estudio (*Daytime* vs *Evening*) mostró diferencias en la permanencia, posiblemente vinculadas a las condiciones laborales, responsabilidades familiares o disponibilidad de recursos de cada grupo. Esto sugiere que las estrategias de retención deben adaptarse a las particularidades de cada modalidad.

### 5. Factores económicos externos (H5)
Las variables macroeconómicas como tasa de desempleo, inflación y PIB mostraron asociación con las tasas de abandono, aunque su efecto es más estructural y menos inmediato que los factores académicos y financieros internos. Esto refuerza la necesidad de políticas institucionales que contemplen la vulnerabilidad económica del entorno.

---

### Síntesis
El riesgo de deserción está fuertemente determinado por el desempeño académico temprano, la situación financiera individual y el acceso a apoyo económico. Factores externos y características de la modalidad también influyen, pero en menor medida y de forma indirecta.

### Recomendaciones

2. Ofrecer **apoyo financiero preventivo** (becas, planes de pago) a estudiantes en riesgo.
3. Diseñar **intervenciones diferenciadas por modalidad** para atender necesidades específicas de estudiantes diurnos y nocturnos.
4. Realizar **monitoreo constante de indicadores macroeconómicos** para anticipar impactos en la retención.

````python
# ============================
# Panel final de gráficas H1–H5
# ============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------- 0) Cargar df si no existe ----------
try:
    df
except NameError:
    # Ajusta esta ruta si tu archivo está en otro lugar
    df = pd.read_csv("/content/drive/MyDrive/desercion_universitaria.csv")

# ---------- 1) Columnas de apoyo y métricas ----------
# Bandera de Dropout
df['Target'] = df['Target'].astype(str).str.strip()
dropout_flag = (df['Target'].str.lower() == 'dropout').astype(int)
df['__dropout_flag__'] = dropout_flag  # columna temporal

# Métricas 1er semestre
if 'pass_rate_1st' not in df.columns:
    enrolled_1st = df['Curricular units 1st sem (enrolled)'].replace(0, np.nan)
    df['pass_rate_1st'] = df['Curricular units 1st sem (approved)'] / enrolled_1st
if 'grade_1st' not in df.columns:
    df['grade_1st'] = df['Curricular units 1st sem (grade)']

# Métricas 2º semestre
if 'pass_rate_2nd' not in df.columns:
    enrolled_2nd = df['Curricular units 2nd sem (enrolled)'].replace(0, np.nan)
    df['pass_rate_2nd'] = df['Curricular units 2nd sem (approved)'] / enrolled_2nd
if 'grade_2nd' not in df.columns:
    df['grade_2nd'] = df['Curricular units 2nd sem (grade)']

# ---------- 2) Bins usados en H1 y H5 ----------
grade_bins = [-np.inf, 10, 12, 14, 16, np.inf]
grade_labels = ['≤10', '10–12', '12–14', '14–16', '>16']
pass_bins   = [-0.01, 0.25, 0.5, 0.75, 1.0]
pass_labels = ['0–25%', '25–50%', '50–75%', '75–100%']

df['grade_1st_bin']     = pd.cut(df['grade_1st'], bins=grade_bins, labels=grade_labels, include_lowest=True)
df['pass_rate_1st_bin'] = pd.cut(df['pass_rate_1st'], bins=pass_bins,   labels=pass_labels)
df['grade_2nd_bin']     = pd.cut(df['grade_2nd'], bins=grade_bins, labels=grade_labels, include_lowest=True)
df['pass_rate_2nd_bin'] = pd.cut(df['pass_rate_2nd'], bins=pass_bins,   labels=pass_labels)

# ---------- 3) Funciones auxiliares ----------
def tasa_dropout_por(col, order=None):
    s = df.groupby(col)['__dropout_flag__'].mean()
    if order is not None:
        s = s.reindex(order)
    return s

def barra(series, titulo, ylabel, xlabel, ylim=(0,1)):
    plt.figure(figsize=(7,5))
    ax = series.plot(kind='bar', edgecolor='black')
    ax.set_title(titulo)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_ylim(*ylim)
    # Anotar valores
    for p in ax.patches:
        h = p.get_height()
        if not np.isnan(h):
            ax.annotate(f"{h:.2f}", (p.get_x()+p.get_width()/2, h),
                        ha='center', va='bottom', fontsize=9)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# ---------- 4) H1: Primer semestre ----------
h1_grade = tasa_dropout_por('grade_1st_bin', grade_labels)
h1_pass  = tasa_dropout_por('pass_rate_1st_bin', pass_labels)

barra(h1_grade, "H1: Dropout por tramo de nota (1º semestre)", "Tasa de Dropout", "Tramo de nota")
barra(h1_pass,  "H1: Dropout por tasa de aprobación (1º semestre)", "Tasa de Dropout", "Tramo de aprobación")

# ---------- 5) H2: Deuda ----------
# Asegurar etiquetas legibles si está codificada binaria
if str(df['Debtor'].dtype).startswith('int') or str(df['Debtor'].dtype).startswith('int8'):
    m = {0: 'No', 1: 'Yes'}
    debtor_plot = df['Debtor'].map(m)
else:
    debtor_plot = df['Debtor']

h2 = pd.concat([debtor_plot, df['__dropout_flag__']], axis=1).groupby('Debtor')['__dropout_flag__'].mean()
barra(h2, "H2: Dropout según deuda", "Tasa de Dropout", "Deuda")

# ---------- 6) H3: Beca ----------
if str(df['Scholarship holder'].dtype).startswith('int'):
    m = {0: 'No', 1: 'Yes'}
    scholar_plot = df['Scholarship holder'].map(m)
else:
    scholar_plot = df['Scholarship holder']

h3 = pd.concat([scholar_plot, df['__dropout_flag__']], axis=1).groupby('Scholarship holder')['__dropout_flag__'].mean()
barra(h3, "H3: Dropout según beca", "Tasa de Dropout", "Beca")

# ---------- 7) H4: Modalidad ----------
# (Daytime/Evening)
h4 = df.groupby('Daytime/evening attendance')['__dropout_flag__'].mean()
barra(h4, "H4: Dropout por modalidad de asistencia", "Tasa de Dropout", "Modalidad")

# ---------- 8) H5: Segundo semestre ----------
h5_pass  = tasa_dropout_por('pass_rate_2nd_bin', pass_labels)
h5_grade = tasa_dropout_por('grade_2nd_bin', grade_labels)

barra(h5_pass,  "H5: Dropout por tasa de aprobación (2º semestre)", "Tasa de Dropout", "Tramo de aprobación")
barra(h5_grade, "H5: Dropout por tramo de nota (2º semestre)", "Tasa de Dropout", "Tramo de nota")

# ---------- 9) Limpieza: quitar columna temporal ----------
df.drop(columns=['__dropout_flag__'], inplace=True, errors='ignore')
````
